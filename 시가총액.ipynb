```python
import pandas as pd
df_2001 = pd.read_csv('csv/2001.csv', encoding='CP949')
df_2001.drop(columns=['종목명','종가','대비','등락률','시가','고가','저가','거래량','거래대금','상장주식수'], inplace=True)
df_2001['연도'] = 2001

list = pd.read_csv('csv/manufacture_list.csv')
list.drop(columns='Unnamed: 0', inplace=True)

for i in list.index:
    if len(str(list.loc[i,"종목코드"])) == 3:
        list.loc[i,"종목코드"]='000'+str(list.loc[i,"종목코드"])
    elif len(str(list.loc[i,"종목코드"])) == 4:
        list.loc[i,"종목코드"]='00'+str(list.loc[i,"종목코드"])
    elif len(str(list.loc[i,"종목코드"])) == 5:
        list.loc[i,"종목코드"]='0'+str(list.loc[i,"종목코드"])
    elif len(str(list.loc[i,"종목코드"])) == 2:
        list.loc[i,"종목코드"]='0000'+str(list.loc[i,"종목코드"])

list_2001 = pd.merge(df_2001, list, on='종목코드')

list_2001 = list_2001.sort_values(by='시가총액', ascending=False).reset_index(drop=True)
list_2001['기업구분'] = 'S'
list_2001['기업구분'].iloc[0:100] = 'B'


list_2001
```
