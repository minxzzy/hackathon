```python
import pandas as pd
```

```python
marcap_2022 = pd.read_csv('csv/2022_marcap.csv', encoding='CP949')
marcap_2022.drop(columns=['종가','대비','등락률','시가','고가','저가','거래량','거래대금','상장주식수'], inplace=True)
marcap_2022['시가총액'] = marcap_2022['시가총액'].sort_values(ascending=False)

marcap_2022
```

```python
pbr_2022 = pd.read_csv('csv/2022_PBR.csv', encoding='CP949')
pbr_2022.drop(columns=['종가','대비','등락률','EPS','PER','선행 EPS','선행 PER','BPS','주당배당금','배당수익률','종목코드'], inplace=True)
pbr_2022 = pbr_2022.sort_values(by=['PBR'], ascending=[True]).reset_index(drop=True)

pbr_2022
```
