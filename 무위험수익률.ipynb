```python
import pandas as pd
rf = pd.read_csv('csv/KOFR.csv')
rf['Date'] = pd.to_datetime(rf['일자'])
rf.drop(columns='일자', inplace=True)
rf = rf.sort_values('Date')
rf.reset_index(drop=True, inplace=True)
rf
```

```python
import yfinance as yf

start_date = '2023-04-01'
end_date = '2024-03-31'

kospi_data = yf.download('^KS11', start=start_date, end=end_date)
```

```python
kospi_data['Date'] = kospi_data.index
kospi_data.reset_index(drop=True, inplace=True)
kospi_data
```

```python
data = pd.merge(rf, kospi_data, on='Date')
data
```

```python
data['rtn'] = data['KOFR'].pct_change() * 100
data['Daily_Return'] = data['Adj Close'].pct_change() * 100

data = data.dropna(axis=0)
data.reset_index(drop=True, inplace=True)

data
```

```python
# kospi_data['Excess_Return'] = kospi_data['Daily_Return'] - rf_sorted['rtn']
result = []

for i in data.index :
    res = data.loc[i]['Daily_Return'] - data.loc[i]['rtn']
    result.append(res)
data['Excess_Return'] = result
data
```
