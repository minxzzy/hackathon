### 1/pbr H / L 처리

```python
pbr_2022 = pd.read_csv('csv/2022_pbr.csv', encoding='CP949')
pbr_2022.drop(columns=['종가','대비','등락률','EPS','PER','선행 EPS','선행 PER','BPS','주당배당금','배당수익률'], inplace=True)
pbr_2022 = pbr_2022.sort_values(by=['PBR'], ascending=[True]).reset_index(drop=True)

pbr_2022['PBR'].fillna(0, inplace=True) # 결측치 0으로 처리


pbr_2022
```

```python
pbr_2022['1/PBR'] = pbr_2022['PBR'].apply(lambda x : 'H' if x < 1 else 'L')
# pbr_2022['1/PBR'].value_counts()
# pbr_2022[pbr_2022['1/PBR'] == 'L']
```

```python
dbi = pd.read_csv('csv/리얼 배당성향.csv', encoding='CP949')
dbi['회사명'] = dbi['회사명'].str.lstrip('(주)')
dbi['회사명'] = dbi['회사명'].str.rstrip('(주)')

dbi.drop(columns=['거래소코드','회계년도'], inplace=True)
dbi.columns = ['종목명','배당성향']

dbi
```

```python
dbi = dbi[(dbi != 0).all(axis=1)]
dbi.dropna(inplace=True)
dbi = dbi.sort_values(by='배당성향', ascending=False).reset_index(drop=True)
dbi
```
