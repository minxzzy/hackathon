```python
import numpy as np

# def weighted_average(group, avg_name, weight_name):
def weighted_average(group, avg_name, _name):
    d = group[avg_name]
    # w = group[weight_name]
    w = group[_name]
    try:
        # return (d * w).sum() / w.sum()
        return d.sum() / w.nunique()
    except ZeroDivisionError:
        return np.nan
    
```

```python
BH_pbr_avgRtn = BH_pbr_rtn.groupby(['Date']).apply(weighted_average, 'sell_rtn', '종목명').to_frame().reset_index().rename(columns={0:'BH_pbr_avg'})
BL_pbr_avgRtn = BL_pbr_rtn.groupby(['Date']).apply(weighted_average, 'sell_rtn', '종목명').to_frame().reset_index().rename(columns={0:'BL_pbr_avg'})
SH_pbr_avgRtn = SH_pbr_rtn.groupby(['Date']).apply(weighted_average, 'sell_rtn', '종목명').to_frame().reset_index().rename(columns={0:'SH_pbr_avg'})
SL_pbr_avgRtn = SL_pbr_rtn.groupby(['Date']).apply(weighted_average, 'sell_rtn', '종목명').to_frame().reset_index().rename(columns={0:'SL_pbr_avg'})
```

```python
from functools import reduce

pbr_avgRtn_list = [BH_pbr_avgRtn, BL_pbr_avgRtn, SH_pbr_avgRtn, SL_pbr_avgRtn]
df_merge = reduce(lambda left, right: pd.merge(left, right, on='Date'), pbr_avgRtn_list)
# pd.merge([BH_pbr_avgRtn, BL_pbr_avgRtn, SH_pbr_avgRtn, SL_pbr_avgRtn], on='Date',axis=1)
df_merge
```

```python
df_merge.iloc[[0]]['BH_pbr_avg']
```

```python
smb = []
hml = []
for i in df_merge.index :
    SMB = ((df_merge.loc[i]['SH_pbr_avg']) + (df_merge.loc[i]['SL_pbr_avg'])) - ((df_merge.loc[i]['BH_pbr_avg']) + (df_merge.loc[i]['BL_pbr_avg'])) / 2
    smb.append(SMB)
    HML = ((df_merge.loc[i]['SH_pbr_avg']) + (df_merge.loc[i]['BH_pbr_avg'])) - ((df_merge.loc[i]['SL_pbr_avg']) + (df_merge.loc[i]['BL_pbr_avg'])) / 2
    hml.append(HML)

df_merge['SMB'] = smb
df_merge['HML'] = hml

df_merge
```
