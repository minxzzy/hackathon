```python
pbr_2022 = pd.read_csv('csv/2022_pbr.csv', encoding='CP949')
pbr_2022.drop(columns=['종목코드','종가','대비','등락률','EPS','PER','선행 EPS','선행 PER','BPS','주당배당금','배당수익률'], inplace=True)
pbr_2022 = pbr_2022.sort_values(by=['PBR'], ascending=[True]).reset_index(drop=True)

pbr_2022['PBR'].fillna(0, inplace=True) # 결측치 0으로 처리


pbr_2022
```

```python
pbr_2022['1/PBR'] = pbr_2022['PBR'].apply(lambda x : 'H' if x < 1 else 'L')
# pbr_2022['1/PBR'].value_counts()
# pbr_2022[pbr_2022['1/PBR'] == 'L']
pbr_2022
```
